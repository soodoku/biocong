## Clean Congress Biography

1. Clean data in `children` column
2. Clean data in `married` column
3. Add a new column for the number of child (son + daughter)
4. Drop the records that have empty biography
5. ?

```python
import re
import pandas as pd
```

```python
pd.options.display.max_colwidth = 500;
pd.options.display.max_rows = 100;
```

```python
df = pd.read_csv('../data/biocong-parsed.csv')
df
```

```python
df.columns
```

### Clean original children column

```python
df.children = df.children.str.strip(';')
```

```python
df['children'] = df.children.str.split(';', n=1, expand=True)[0]
df
```

### Clean married column

```python
df['married'] = df.married.str.split(';', n=1, expand=True)[0]
df
```

### Extract number of son, daughter and child

```python
def extract_children(bio, kw='child'):
    if not pd.isnull(bio):
        #print(bio)
        for m in re.finditer('[;,:](\s\w+\s%s)' % kw, bio):
            return m.group(1)
    return None
```

```python
df['n_daughter'] = df.biography.apply(lambda c: extract_children(c, 'daughters?'))
df['n_son'] = df.biography.apply(lambda c: extract_children(c, 'sons?'))
df['n_children'] = df.biography.apply(lambda c: extract_children(c, 'child(:?ren)?'))
df
```

### Combine Children/Son/Daughter to a column

```python
df['n_children'] = df[['n_children', 'n_son', 'n_daughter']].apply(lambda x: ','.join([a for a in x if a]), axis=1)
del df['n_son']
del df['n_daughter']
df
```

###  Show the raw HTML file of the records that have empty biography

```python
for i, row in df.iterrows():
    if pd.isnull(row.biography):
        with open('../data/' + row.textfile) as f:
            html = f.read()
            m = re.match('.*<pre>(.*)</pre>', html, re.S)
            if m:
                text = m.group(1)
                print(row.textfile, text)
```

### Drop empty biography records

```python
df.dropna(subset=['biography'], inplace=True)
df
```

### Show the number of records if drop duplicate by name

```python
df.drop_duplicates(subset=['name'])
```

### Show the number of records if drop duplicate by biography

```python
df.drop_duplicates(subset=['biography'])
```

### Save clean output

```python
df.to_csv('../data/biocong-cleaned.csv', index=False)
```
