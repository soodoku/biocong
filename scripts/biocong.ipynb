```python
import requests
import pandas as pd
```

```python
def crawl_biocong(path, records=[]):
    r = requests.get(path)
    code = r.status_code
    if code != 200:
        print(code, path)
    for c in r.json()['childNodes']:
        value = c['nodeValue'].get('value', None)
        #disp_value = c['nodeValue'].get('displayValue', None)
        path = c['nodeValue'].get('browsePath', None)
        #print(path, value, disp_value)
        if value:
            records = crawl_biocong('https://www.govinfo.gov/wssearch/rb//cdir/%s?fetchChildrenOnly=1' % path, records)
        else:
            if path.endswith('CONGRESSMEMBERSTATE'):
                # Interesting fields:- title, section, browseline1, firstPage, publishdate, textfile
                records.append(c['nodeValue'])
        #break
    print('Number of records:', len(records))        
    return records
```

```python
records = crawl_biocong('https://www.govinfo.gov/wssearch/rb/cdir?fetchChildrenOnly=0')
```

```python
import os


def download_file(url, folder_name):
    local_filename = url.split('/')[-1]
    path = os.path.join("{}/{}".format(folder_name, local_filename))
    with requests.get(url, stream=True) as r:
        with open(path, 'wb') as f: 
            for chunk in r.iter_content(chunk_size=1024):
                if chunk:
                    f.write(chunk)

    return local_filename
```
