```python
import pandas as pd
```

```python
df = pd.read_csv('../data/biocong-cleaned-parse-name.csv')
df
```

```python
df.columns
```

```python
df.children
```

```python
import re
from nameparser import HumanName
```

```python
def split_children(s):
    out = []
    if s is None or pd.isnull(s):
        return pd.Series(dtype=object)
    for n in re.split('\,(?!(?:JR|SR|PHD[^\.]))|and\s+', s, flags=re.I):
        n = n.strip()
        if n != '':
            name = HumanName(n)
            if name.first:
                out.append(name.first)
    if len(out):
        return pd.Series(out)
    else:
        return pd.Series(dtype=object)
```

```python
ndf = df.children.apply(lambda c: split_children(c))
ndf
```

```python
cols = ['child_name%s' % col for col in ndf.columns]
ndf.columns = cols
ndf
```

```python
odf = df.join(ndf)
odf
```

```python
odf.to_csv('../data/biocong-cleaned-parse-name+children.csv', index=False)
```
